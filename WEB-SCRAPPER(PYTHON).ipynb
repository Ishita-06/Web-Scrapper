```python
import requests
from bs4 import BeautifulSoup as bs   #this library is used to beautify the data so that it can be readable by user
from urllib.request import urlopen as uReq
import logging
```

```python
flipkart_url="https://www.flipkart.com/search?q="+"iPhone11"
```

```python
flipkart_url
```

```python
urlclient=uReq(flipkart_url)
```

```python
urlclient
```

```python
flipkart_page=urlclient.read()  #it reads all the information available in the web page that the url will direct you into
```

```python
flipkart_html=bs(flipkart_page,'html.parser')
```
